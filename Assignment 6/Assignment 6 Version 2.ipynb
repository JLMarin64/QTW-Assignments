```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
from sklearn.metrics import roc_auc_score
import pandas as pd
import numpy as np
```

```python
N=500000. #Change this line adjust the number of rows. 
data=pd.read_csv("HIGGS.csv",nrows=N,header=None)
test_data=pd.read_csv("HIGGS.csv",nrows=10000,header=None,skiprows=10500000)

from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
from sklearn.metrics import roc_auc_score

```

# 1.	Pick 3 or more different architectures (add/subtract layers+neurons) and run the model + score. 

```python
y = np.array(data.loc[:,0])
x = np.array(data.loc[:,1:])
x_test = np.array(test_data.loc[:,1:])
y_test = np.array(test_data.loc[:,0])

#Begin 

model = Sequential()
model.add(Dense(50, input_dim=x.shape[1], kernel_initializer='uniform')) # X_train.shape[1] == 28 here
model.add(Activation('sigmoid'))
model.add(Dropout(0.10))
model.add(Dense(50, kernel_initializer='uniform'))
model.add(Activation('sigmoid'))
model.add(Dropout(0.10))
model.add(Dense(1, kernel_initializer='uniform')) 
model.add(Activation('sigmoid'))

sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer=sgd)

model.fit(x, y, epochs=1000, batch_size=1000)
roc_auc_score(y_test,model.predict(x_test))
#end
```

# Q1: What was the effect of adding more layers/neurons.

When adding layers/neurons the accuracy does not have a wide fluctuation for the three architectures that have been estimated. The base model for comparsion is 3 layers with 50 neurons in each layer (accuracy = 0.8265). When the number of layers increased to 4 and the number of neurons increased to 100 for each layer, the accuracy (0.8144) is lower than the base model. When decreasing the number of neurons to 25, the accuracy (0.8184) is lower than the base model but higher than the 4 layer and 100 neuron model. 

# Q2: Which parameters gave you the best result and why (in your opinion) did they work.

The parameters that worked is using He_Uniform kernel initializer, Sigmoid activation, and Nadam optimization. The He_Uniform kernel initializer is able to create weights of the layer closer to the "true" weight matrix. The main difference between He_Uniform and the Uniform kernel initializer is that the range to which the intial random weight matrix is created is wider for He_Uniform than the Uniform. Therefore, it seems the "true" weight matrix is outside the range of Uniform. Also using Sigmoid activation seems to be the best, especially, for classifiers. Because of the S shape, bounded between 0 to 1 it seems ideal for this dataset. Tanh is also used since it follows this same shape, however, the accuracy is horrible indicating that it may fall out of the range for the function. The Nadam optimizer seems to be the best since it utilizes the Nesterov momentum, which is like a "corrector" for the gradient descent. The optimizer also uses preivous gradients, which are utilized in the calculation of the current gradient descent. Therefore, able to converge to the minimum faster for the loss function. Also from the previous runs adding layers and decreasing neurons show a drop in accuracy. This seems to indicate that layers does not increase accuracy, but maybe the current or higher neuron count may increase the accuracy since there is more neurons to learn the data. Also having a smaller batch size seems to increase accuracy but increased run time. The smaller batch size is able to have the model "learn" more from the data, and therefore not have the model miss any important information. From the testing, increasing batch decreased accuracy but decreasing batch size increased accuracy.
